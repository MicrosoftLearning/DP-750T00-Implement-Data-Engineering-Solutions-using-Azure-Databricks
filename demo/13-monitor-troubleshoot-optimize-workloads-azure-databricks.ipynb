## Track costs with budgets and system tables

![diagram](https://raw.githubusercontent.com/MicrosoftLearning/DP-750T00-Implement-Data-Engineering-Solutions-using-Azure-Databricks/refs/heads/main/demo/mindmaps/learn.wwl.monitor-troubleshoot-optimize-workloads-azure-databricks/02-05-track-costs-budgets.svg)

````python
-- Daily DBU consumption trend
SELECT
  usage_date,
  sum(usage_quantity) as DBUs_Consumed
FROM system.billing.usage
WHERE sku_name = "STANDARD_ALL_PURPOSE_COMPUTE"
GROUP BY usage_date
ORDER BY usage_date ASC
````

````python
-- Jobs consuming most DBUs
SELECT
  usage_metadata.job_id as job_id,
  sum(usage_quantity) as total_dbus
FROM system.billing.usage
WHERE usage_metadata.job_id IS NOT NULL
GROUP BY job_id
ORDER BY total_dbus DESC
````

````python
-- Cost by cluster owner
SELECT
  c.owned_by,
  sum(u.usage_quantity) as total_dbus
FROM system.billing.usage u
JOIN system.compute.clusters c
  ON u.usage_metadata.cluster_id = c.cluster_id
WHERE u.usage_metadata.cluster_id IS NOT NULL
GROUP BY c.owned_by
ORDER BY total_dbus DESC
````

````python
-- Usage by custom tag
SELECT
  sku_name,
  usage_unit,
  sum(usage_quantity) as total_usage
FROM system.billing.usage
WHERE custom_tags['project'] = 'sales-analytics'
GROUP BY sku_name, usage_unit
````

## Investigate data skew

![diagram](https://raw.githubusercontent.com/MicrosoftLearning/DP-750T00-Implement-Data-Engineering-Solutions-using-Azure-Databricks/refs/heads/main/demo/mindmaps/learn.wwl.monitor-troubleshoot-optimize-workloads-azure-databricks/05-03-data-skew.svg)

````python
spark.conf.get("spark.databricks.optimizer.adaptive.enabled")
````

````python
from pyspark.sql.functions import spark_partition_id, count

df = spark.read.table("samples.nyctaxi.trips")
partition_counts = df.withColumn("partition", spark_partition_id()) \
    .groupBy("partition") \
    .agg(count("*").alias("row_count")) \
    .orderBy("partition")
partition_counts.show()
````

````python
from pyspark.sql.functions import col, concat, lit, floor, rand

df_large = spark.read.table("samples.nyctaxi.trips")
num_salts = 10
df_salted = df_large.withColumn(
    "salted_key",
    concat(col("pickup_zip"), lit("_"), (floor(rand() * num_salts)).cast("string"))
)
print("Salted key sample:")
df_salted.select("pickup_zip", "salted_key").show(5)
````

## Investigate shuffle issues

![diagram](https://raw.githubusercontent.com/MicrosoftLearning/DP-750T00-Implement-Data-Engineering-Solutions-using-Azure-Databricks/refs/heads/main/demo/mindmaps/learn.wwl.monitor-troubleshoot-optimize-workloads-azure-databricks/05-05-shuffle-issues.svg)

````python
from pyspark.sql.functions import broadcast

trips = spark.read.table("samples.nyctaxi.trips")
small_lookup = spark.createDataFrame([
    ("10001", "Manhattan"),
    ("10002", "Brooklyn"),
    ("10003", "Queens")
], ["zip", "borough"])

print("=== Regular join plan ===")
trips.join(small_lookup, trips.pickup_zip == small_lookup.zip).explain()

print("\n=== Broadcast join plan ===")
trips.join(broadcast(small_lookup), trips.pickup_zip == small_lookup.zip).explain()
````

## Query logs with Kusto Query Language

![diagram](https://raw.githubusercontent.com/MicrosoftLearning/DP-750T00-Implement-Data-Engineering-Solutions-using-Azure-Databricks/refs/heads/main/demo/mindmaps/learn.wwl.monitor-troubleshoot-optimize-workloads-azure-databricks/06-03-query-logs-kql.svg)

```kusto
// Recent job events (last 24 hours)
DatabricksJobs
| where TimeGenerated > ago(24h)
| project TimeGenerated, OperationName, Identity, SourceIPAddress
| order by TimeGenerated desc
| take 100
```

```kusto
// Failed job runs (last 7 days)
DatabricksJobs
| where TimeGenerated > ago(7d)
| where Response contains "error" or Response contains "failed"
| summarize FailureCount = count() by OperationName, bin(TimeGenerated, 1h)
| order by FailureCount desc
```

```kusto
// Cluster usage patterns
DatabricksClusters
| where TimeGenerated > ago(7d)
| summarize EventCount = count() by OperationName
| order by EventCount desc
```

```kusto
// Monitor Unity Catalog access
DatabricksUnityCatalog
| where TimeGenerated > ago(24h)
| where OperationName has "getTable" or OperationName has "selectFromTable"
| project TimeGenerated, Identity, OperationName, RequestParams
| order by TimeGenerated desc
```

## Create alerts for proactive monitoring

![diagram](https://raw.githubusercontent.com/MicrosoftLearning/DP-750T00-Implement-Data-Engineering-Solutions-using-Azure-Databricks/refs/heads/main/demo/mindmaps/learn.wwl.monitor-troubleshoot-optimize-workloads-azure-databricks/06-04-create-alerts.svg)

```kusto
// Alert: Job failures > 5 in last hour
DatabricksJobs
| where TimeGenerated > ago(1h)
| where Response contains "error" or Response contains "failed"
| summarize FailureCount = count()
| where FailureCount > 5
```

```kusto
// Alert: Unity Catalog access outside business hours
DatabricksUnityCatalog
| where TimeGenerated > ago(1h)
| where hourofday(TimeGenerated) < 6 or hourofday(TimeGenerated) > 20
| summarize AccessCount = count() by Identity
| where AccessCount > 10
```
